```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.linear_model import Ridge, Lasso
```

```python
df = pd.read_csv('session2_task.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
Scaler = StandardScaler()
X_train_scaled = Scaler.fit_transform(X_train)
X_test_scaled = Scaler.fit_transform(X_test)
```

```python
poly= PolynomialFeatures(degree= 2)
X_train_poly= poly.fit_transform(X_train_scaled)
X_test_poly= poly.fit_transform(X_test_scaled)
```

```python
# # Ridge Regression
# ridge = Ridge(alpha=0.1)
# ridge.fit(X_train_poly, y_train)
# y_pred_ridge = ridge.predict(poly.fit_transform(X_test))
# print("Ridge Regression R2 Score:", r2_score(y_test, y_pred_ridge))

# Lasso Regression
lasso = Lasso(alpha=0.1)
lasso.fit(X_train_poly, y_train)
y_pred_lasso = lasso.predict(X_test_poly)
print("Lasso Regression R2 Score:", r2_score(y_test, y_pred_lasso))
```
